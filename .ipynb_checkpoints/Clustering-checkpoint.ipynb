## Метод k-средних

```python
# Импортируем библиотеки
from sklearn import datasets
from sklearn.cluster import KMeans

# Загружаем набор данных
iris_df = datasets.load_iris()

# Описываем модель
kmeans = KMeans(n_clusters=3, random_state=13)

# Проводим моделирование
kmeans.fit(iris_df.data)

# Предсказание на единичном примере
predicted_label = kmeans.predict([[7.2, 3.5, 0.8, 1.6]])

# Предсказание на всем наборе данных
all_predictions = kmeans.predict(iris_df.data)

# Выводим предсказания
print(predicted_label)
print(all_predictions)
```

```python
# Результат кластеризации (на примере зависимость между длиной и шириной чашелистика)
# Так как k-means случайно нумерует кластеры, то можно их самостоятельно преобразовать для визуализации
def irm(x):
    if x==0: return 1
    if x==1: return 0
    if x==2: return 2
plt.scatter(x_axis, y_axis, c=list(map(irm, all_predictions)))
```

```python
plt.scatter(x_axis, y_axis, c=iris_df.target)
```

### Оценка выбора количества кластеров

```python
inertia = []
for k in range(1, 8):
    kmeans_iter = KMeans(n_clusters=k, random_state=1).fit(iris_df.data)
    inertia.append(np.sqrt(kmeans_iter.inertia_))

plt.plot(range(1, 8), inertia, marker='s');
plt.xlabel("Число кластеров")
plt.ylabel("Cумма квадратов расстояний\nот точек до центроидов");
```

## Метод кластеризации на основе плотности DBSCAN и PCA

```python
# Импортируем библиотеки
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA

# Загружаем датасет
iris = load_iris()

# Определяем модель
dbscan = DBSCAN()

# Обучаем
dbscan.fit(iris.data)

# Уменьшаем размерность при помощи метода главных компонент
pca = PCA(n_components=2).fit(iris.data)
pca_2d = pca.transform(iris.data)

# Строим в соответствии с тремя классами
for i in range(0, pca_2d.shape[0]):
    if dbscan.labels_[i] == 0:
        c1 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='r', marker='+')
    elif dbscan.labels_[i] == 1:
        c2 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='g', marker='o')
    elif dbscan.labels_[i] == -1:
        c3 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='b', marker='*')

plt.legend([c1, c2, c3], ['Кластер 1', 'Кластер 2', 'Шум'])
plt.title('DBSCAN нашел 2 кластера и шум')
plt.show()

```

```python
plt.scatter(pca_2d[:, 0], pca_2d[:, 1], c=iris_df.target)
```

### Применение k-means после обработки PCA

```python
# Проводим моделирование
kmeans.fit(pca_2d)

# Предсказание на всем наборе данных
all_predictions_pca = kmeans.predict(pca_2d)
plt.scatter(pca_2d[:, 0], pca_2d[:, 1], c=all_predictions_pca)
```

### DBSCAN с настройкой параметров eps и min_samples (пример кода для визуализации из sklearn)

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

X = iris.data
labels_true = iris.target
# Compute DBSCAN
db = DBSCAN(eps=0.5, min_samples=5).fit(iris.data)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print('Estimated number of clusters: %d' % n_clusters_)
print('Estimated number of noise points: %d' % n_noise_)
# print("Homogeneity: %0.3f" % metrics.homogeneity_score(labels_true, labels))
# print("Completeness: %0.3f" % metrics.completeness_score(labels_true, labels))
# print("V-measure: %0.3f" % metrics.v_measure_score(labels_true, labels))
# print("Adjusted Rand Index: %0.3f"
#       % metrics.adjusted_rand_score(labels_true, labels))
# print("Adjusted Mutual Information: %0.3f"
#       % metrics.adjusted_mutual_info_score(labels_true, labels))
# print("Silhouette Coefficient: %0.3f"
#       % metrics.silhouette_score(X, labels))

# #############################################################################
# Plot result
import matplotlib.pyplot as plt

# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```

## Понижение размерности с методом t-SNE

```python
# Импорт библиотек
from sklearn import datasets
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Загрузка датасета
iris_df = datasets.load_iris()

# Определяем модель и скорость обучения
tsne1 = TSNE(learning_rate=100, random_state=13)

# Обучаем модель
transformed1 = tsne1.fit_transform(iris_df.data)

# Представляем результат в двумерных координатах
x_axist1 = transformed1[:, 0]
y_axist1 = transformed1[:, 1]

plt.scatter(x_axist1, y_axist1, c=iris_df.target)
plt.show()
```

### Различные параметры random_state могут дать разный результат

```python
# Определяем модель и скорость обучения
tsne2 = TSNE(learning_rate=100, random_state=39)

# Обучаем модель
transformed2 = tsne2.fit_transform(iris_df.data)

# Представляем результат в двумерных координатах
x_axist3 = transformed2[:, 0]
y_axist3 = transformed2[:, 1]

plt.scatter(x_axist3, y_axist3, c=iris_df.target)
plt.show()
```

### Применение k-means после обработки t-SNE

```python
# Проводим моделирование
kmeans.fit(transformed1)

# Предсказание на всем наборе данных
all_predictions_pca = kmeans.predict(transformed1)
plt.scatter(transformed1[:, 0], transformed1[:, 1], c=list(map(irm, all_predictions_pca)))
```

### Различные значения параметра `perplexity` могут дать существенно разные результаты

```python
# Определяем модель и скорость обучения
tsne3 = TSNE(n_components=2,learning_rate=100, perplexity=10, random_state=13)

scaler = StandardScaler()
X = scaler.fit_transform(iris_df.data)
# Обучаем модель
transformed3 =tsne3.fit_transform(X)

# Представляем результат в двумерных координатах
x_axist4 = transformed3[:, 0]
y_axist4 = transformed3[:, 1]

plt.scatter(x_axist4, y_axist4, c=iris_df.target)
plt.show()
```

## Иерархическая кластеризация и построение дендрограммы

```python
# Импортируем библиотеки
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import pandas as pd

# Создаем датафрейм
seeds_df = pd.read_csv("http://qps.ru/jNZUT")

# Исключаем информацию об образцах зерна, сохраняем для дальнейшего использования
varieties = list(seeds_df.pop('grain_variety'))

# Извлекаем измерения как массив NumPy
samples = seeds_df.values

# Реализация иерархической кластеризации при помощи функции linkage
mergings = linkage(samples, method='complete')

# Строим дендрограмму, указав параметры удобные для отображения
dendrogram(mergings,
           labels=varieties,
           leaf_rotation=90,
           leaf_font_size=6,
           )

plt.show()

```
